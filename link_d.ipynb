```python
import pandas as pd
import requests
import os
import mimetypes


file_name = "cardboard box waste - Sheet1.csv"
df = pd.read_csv(file_name)


print("Preview of data:")
print(df.head())

print("\nColumns:")
print(df.columns)


column_name = "image"

df["clean_url"] = df[column_name].astype(str).str.replace("🔗", "", regex=False)

print("\nClean URLs preview:")
print(df["clean_url"].head())


output_folder = "downloaded_images"
os.makedirs(output_folder, exist_ok=True)


limit = 1000

for i, url in enumerate(df["clean_url"][:limit]):
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()

        content_type = response.headers.get('content-type')
        extension = mimetypes.guess_extension(content_type) or ".jpg"

        file_path = os.path.join(output_folder, f"image_{i}{extension}")

        with open(file_path, "wb") as f:
            f.write(response.content)

        print(f"Downloaded: {file_path}")

    except Exception as e:
        print(f"Error downloading {url}: {e}")

print("\n Done downloading images")
```
